```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge , Lasso
from sklearn.metrics import mean_squared_error , r2_score 
from sklearn.datasets import load_diabetes
```

```python
df = pd.read_csv(r"C:\Users\air73\OneDrive\Documents\Full stack data science\diabetes.csv")
```

```python
x = df[["Pregnancies","Glucose",	"BloodPressure", "SkinThickness" , "Insulin",	"BMI",	"DiabetesPedigreeFunction",	"Age"]]
```

```python
y = df[["Outcome"]]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
ridge_model = Ridge()
lasso_model = Lasso()
```

```python
#creating a ridge model 
rideg_model = ridge_model.fit(x_train_scaled , y_train)
ridge_model
```

```python
lasso_model = lasso_model.fit(x_train_scaled , y_train)
lasso_model
```

```python
#predicting the data from ridge model 
ridge_model.predict(x_test_scaled)
```

```python
#predicting the data from lasso model 
lasso_model.predict(x_test_scaled)
```

```python
#Calculating the error in ridge model
mean_squared_error(y_test , ridge_model.predict(x_test_scaled))
```

```python
#calculating the error for lasso model 
mean_squared_error(y_test ,lasso_model.predict(x_test_scaled))
```

```python
# Calculating the R-Square for lasso model 
r2_lasso = r2_score(y_test , lasso_model.predict(x_test_scaled))            # y_test == y_true and model.predict(x_test_scaled)== y_predict
r2_lasso
```

```python
# Calculating the R-Square for Ridge model 
r2_ridge = r2_score(y_test , ridge_model.predict(x_test_scaled))
r2_ridge
```

```python
# Calculating the adjusted R-Square for lasso model
n = len(y_test)      # No of data points/rows
p = x.shape[1]       # No of predictors(independent variables i.e age , height , weight , gender) / coloumns
adj_r2_lasso = 1 - (1-r2_lasso) * (n-1) / (n-p-1)
adj_r2_lasso
```

```python
# Calculating the adjusted R-Square for ridge model 
adj_r2_ridge = 1 - (1-r2_ridge) * (n-1) / (n-p-1)
adj_r2_ridge
```

```python
# Calculating the RMSE for lasso model 
rmse_lasso = np.sqrt(mean_squared_error(y_test , lasso_model.predict(x_test_scaled)))
rmse_lasso
```

```python
# Calculating the RMSE for ridge model 
rmse_ridge = np.sqrt(mean_squared_error(y_test , ridge_model.predict(x_test_scaled)))
rmse_ridge
```
